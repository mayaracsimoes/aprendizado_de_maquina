2) Utilize a base de dados maior (mercado2) que apresenta o histórico de compras de um supermercado francês em 1 semana. Teste o seu algoritmo e responda as seguintes perguntas:

```python
import pandas as pd
from apyori import apriori

base_mercado = pd.read_csv('data/mercado2.csv', header = None)
transacoes = []
qtd_colunas = base_mercado.shape[1]
for i in range(0, len(base_mercado)):
     transacoes.append([str(base_mercado.values[i, j]) for j in range(qtd_colunas) if str(base_mercado.values[i, j]) != 'nan'])
```

a) Vamos supor que você quer extrair Regras da Associação para os produtos que são vendidos pelo menos quatro vezes ao dia, ou seja, 28 vezes por semana (4*7). Adicione o valor de suporte de acordo com o problema.

Para adicionar o valor de suporte, você deve calcular o valor de 28 (quantidade de produtos vendidos) dividido pela quantidade total de produtos na base de dados.

Execute o algoritmo e responda qual é o resultado.

```python
suporte_minimo = 28 / len(transacoes)

print(f"Suporte mínimo calculado: {suporte_minimo:.4f}")

regras = apriori(transacoes, min_support=suporte_minimo)
resultados = list(regras)
print(f"Total de regras geradas: {len(resultados)}")
```

Resultado:
Suporte mínimo calculado: 0.0037
Total de regras geradas: 1104

b) Adicione o valor de confiança para 0.2 e o lift para 3. Quantos registros foram retornados?

```python
regras = apriori(transacoes, min_support=suporte_minimo, min_confidence=0.2, min_lift= 3)
resultados = list(regras)
print(f"Total de regras geradas: {len(resultados)}")
```

Total de regras geradas: 44

c) Adicione o código abaixo para poder visualizar melhor os dados. Faça comentário do que acontece em cada linha do código.

```python
A = []
B = []
suporte = []
confianca = []
lift = []

# Percorre todas as regras encontradas
for resultado in resultados:
    s = resultado[1]            # Valor do suporte da regra
    result_rules = resultado[2] # Lista com as relações entre itens (se → então)

    # Percorre regra dentro do resultado
    for result_rule in result_rules:
        a = list(result_rule[0])   # Se (parte A da regra)
        b = list(result_rule[1])   # Então (parte B da regra)
        c = result_rule[2]         # Confiança
        l = result_rule[3]         # Lift

        # Armazena cada valor nas listas
        A.append(a)
        B.append(b)
        suporte.append(s)
        confianca.append(c)
        lift.append(l)

# Cria um DataFrame com todas as informações das regras
rules_df = pd.DataFrame({
    'A': A,
    'B': B,
    'suporte': suporte,
    'confianca': confianca,
    'lift': lift
})

# Ordena as regras pelo valor de lift
rules_df.sort_values(by='lift', ascending=False)
```
